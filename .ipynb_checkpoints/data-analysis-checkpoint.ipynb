```python
import pandas as pd 
import numpy as  np 
import matplotlib.pyplot as plt 
import seaborn as sns
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
df =pd.read_csv('movies.csv')
```

```python
selected_features = ['genres','keywords','tagline','cast','director']
```

```python
# repalacing the null valuess with null string 

for feature in selected_features : 
    df[feature] = df[feature].fillna('')
```

```python
combained_features = df['genres']+" "+df['keywords']+" "+df['tagline']+" "+df['cast']+" "+df['director']
```

```python
combained_features
```

## Convert the text data to features vectors data 

```python
vectorizer =TfidfVectorizer() 
```

```python
feature_vectors =vectorizer.fit_transform(combained_features) 
```

```python
print(feature_vectors)
```

## Cosine Similarity

```python
# getting the simliarity scores using cosine similarity  


similarity =cosine_similarity(feature_vectors) 
```

```python
similarity
```

```python
## getting the movie name  
movies_name =input('Enter you Favourite Movie Name : ') 
```

```python
list_of_title = df['title'].tolist() 

print(list_of_title)
```

```python
## finding closed match for the movies name given by the user 
find_close_match =difflib.get_close_matches(movies_name,list_of_title) 

print(find_close_match)
```

```python
closed_match  =find_close_match[0]
```

```python
print(closed_match)
```

```python
# finding the index of the movie with title 
index_of_movie = df[df.title == closed_match]['index'].values[0]
print(index_of_movie)
```

```python
## getting list of similar movies
similarity_score =list(enumerate(similarity[index_of_movie]))  
```

```python
similarity_score
```

```python
len(similarity_score)
```

```python
sorted_similarity =sorted(similarity_score,key=lambda x:x[1],reverse=True) 
```

```python
sorted_similarity
```

```python
# print the name of similar movies based on the index

print('Movie suggested for you  :\n') 
i = 1
for movie in sorted_similarity: 
    indx =movie[0]  
    title_from_index =df[df.index == indx]['title'].values[0] 
    if i < 31 : 
        print(i, '.',title_from_index) 
        i+=1 
```
